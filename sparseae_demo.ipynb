## Sparse Autoencoder Demo

### About

Sparse autoencoders are a way of automatically learning features from unlabeled data. Images are one such plentiful data source. This notebook demonstrates using a basic autoencoder to extract features from natural black and white images. 

This python implementation follows the exercise in the [UFLDL tutorial](http://ufldl.stanford.edu/wiki/index.php/Exercise:Sparse_Autoencoder).

All code is available [here](https://github.com/dbolotov/learning_deep_learning).

The network architecture consists of one hidden layer between the input and output layers. The input and output layers are the same, and the main idea is that due to certain constraints on the hidden layer, the network is forced to learn a compressed representation of the data. The two constraints used here are choosing a low number of hidden units and forcing the average activations of the those units to be close to zero. A detailed explanation is available on this [UFLDL page](http://nlp.stanford.edu/~socherr/sparseAutoencoder_2011new.pdf).

```python
from sparseae_functions import *
from sparseae import *
```

We start from a set of black and white images of natural scenery:

Sample some random 8 pixel by 8 pixel patches from the set, and show a few:

```python
patches = sample_images(numpatches=10000,patchsize=8)
display_network(patches)
```

The data is preprocessed before learning. Normalizing the images includes subtracting the mean, truncating to +/- 3 standard deviations, and scaling to `[0.1,0.9]`. The same patches are shown after normalizing:

```python
patches = normalize_data(patches)
display_network(patches)
```

The data is now ready for training. Now we initialize a network (using default parameters). Since we're using 8x8 images to train, the input vector size is 64. The hidden layer here is chosen to consist of 25 features. This implementation uses sigmoid activation functions.

```python
sae = sparseae()
print sae.vSize #size of input and output layers
print sae.hSize #size of hidden layer
```

We now train the network. This function is using an L-BFGS implementation in `scipy.optimize`:

```python
weights = sae.train(patches)
```

```python
display_weights(arr = weights, vSide = sae.vSide, hSide = sae.hSide)
```

Looking at the trained weights, we see that the network learned some edges!
